General Thumb Rules:

* For features with less missing values- can use regression to predict the missing values or fill with the mean of the values present, depending on the feature.
* For features with very high number of missing values- it is better to drop those columns as they give very less insight on analysis.
* As there's no thumb rule on what criteria do we delete the columns with high number of missing values, but generally you can delete the columns, if you have more than 30-40% of missing values. But again there's a catch here, for example, Is_Car & Car_Type, People having no cars, will obviously have Car_Type as NaN (null), but that doesn't make this column useless, so decisions has to be taken wisely.

# Data Cleaning

#### 1. Create a copy of base dataset for manupulation and preprocessing

```python
data_new = data.copy()
```

```python
data_new.head(5)
```

#### 2. TotalCharges should be numeric amount. let's convert it to numerical datatype

```python
data_new.TotalCharges = pd.to_numeric(data_new.TotalCharges, errors = "coerce")
data_new.isnull().sum()
```

#### 3. As we can see there are 11 missing values in TotalCharges column. Let's check these records.

```python
data_new.loc[data_new["TotalCharges"].isnull()== True]
```

#### 4. Missing values Treatment

- Since the % of these records compared to total dataset is very low ie 0.15%, it is safe to ignore them from further processing.

```python
# Removing missing values
data_new.dropna(how = "any", inplace = True)
```

#### 5. Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...

```python
print(data_new["tenure"].max())
```

```python
# Group the tenure in binns of 12 months format

labels = ["{0}- {1}".format(i,i+11)for i in range(1,72,12)]

data_new["Tenure Group"] = pd.cut(data_new.tenure ,range(1,80,12), right = False , labels = labels)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def uniplot(df, col, title=None, hue=None, bins=20):
    """
    Universal plotting function for single-column analysis.
    
    Parameters:
    df   : pandas DataFrame
    col  : Column name to plot
    title: Plot title (optional)
    hue  : Column name for grouping (optional)
    bins : Number of bins for numeric histograms (default=20)
    """
    
    # Styling
    sns.set(style='whitegrid', context='talk')
    plt.rcParams.update({
        "axes.labelsize": 16,
        "axes.titlesize": 18,
        "axes.titlepad": 20
    })

    # Title default
    if not title:
        title = f"Distribution of {col}"

    # Dynamic figure width
    hue_unique = df[hue].nunique() if hue else 0
    fig_width = len(df[col].unique()) + 7 + 4 * hue_unique if df[col].dtype == 'object' or df[col].nunique() < 25 else 10
    fig, ax = plt.subplots(figsize=(fig_width, 6))

    # Choose plot based on data type
    if pd.api.types.is_numeric_dtype(df[col]):
        sns.histplot(
            data=df,
            x=col,
            hue=hue,
            bins=bins,
            kde=True,
            palette='bright',
            ax=ax
        )
    else:
        sns.countplot(
            data=df,
            x=col,
            hue=hue,
            order=df[col].value_counts().index,
            palette='bright',
            ax=ax
        )
        plt.xticks(rotation=45)

    ax.set_title(title)
    plt.tight_layout()
    plt.show()
```
